```python
#Ejericicio de depreciacion usando el metodo lineal

C = 100000
N = 4

#Da[N] = 80000 || Da[N] = B

B = 80000

print("Costo del activo: ",C)
print("Vida util del activo: ",N)
print("Base de depreciacion: ",B)
```

```python
#Calculamos el salvamento

S = C - B

print("Valor de salvamento del activo: ",S)
```

```python
#Calcular la depreciacion Dt[2] y Dt[3]

Dt = B/N

print("Depreciacion del activo: ", Dt)
```

```python
#Calcular depreciacion acumulada

Da = [0]
Vt = [0]
Vt[0] = C
for i in range(1,N+1):
    Da.append(i*Dt)
    Vt.append(C-Da[i])
```

```python
from tabulate import tabulate

#imprimimos la tabla
table = []
for i in range(0, N+1):
    table.append([i, Dt, Da[i], Vt[i]])

print(tabulate(table, headers=["Año", "Depreciacion", "Depreciacion acumulada", "Valor del activo"], tablefmt="fancy_grid"))
```

```python
#Graficamos la tabla
import matplotlib.pyplot as plt

plt.plot(Dt, label="Depreciacion")
plt.plot(Da, label="Depreciacion acumulada")
plt.plot(Vt, label="Valor del activo")
plt.legend()
plt.show()
```
